- **Центральный процессор (ЦП)**: Основной элемент, выполняющий вычисления и обработку данных.

- **Аудио**: Звук через динамики или наушники.
- **Печать**: Вывод информации на бумаге через принтер.
- **Сохранение данных**: Запись информации на внешние носители или в облако.

#### 4. Взаимодействие с окружающей средой (Environment Interaction)
- **Пользователь**: Взаимодействует с ПК через интерфейсы (графические, текстовые).

- **Внешние устройства**: Подключаемые устройства, такие как USB-накопители, внешние жесткие диски, камеры и т.д.

### Модель состава системы "Персональный компьютер"

#### 1. Аппаратное обеспечение (Hardware)
- **Процессор (ЦП)**: Центральный элемент, выполняющий вычисления и управление.
- **Материнская плата**: Основная плата, на которой размещаются все компоненты и обеспечивается их взаимодействие.
- **Оперативная память (ОП)**: Временное хранилище данных и программ, которые активно используются.
- **Жесткий диск (HDD) / Твердотельный накопитель (SSD)**: Устройства для долговременного хранения данных.
- **Графический процессор (ГП)**: Обрабатывает графику и видео, особенно важен для игр и мультимедийных приложений.
- **Блок питания**: Обеспечивает электроэнергией все компоненты системы.
- **Устройства ввода**: Клавиатура, мышь, сканеры и другие устройства для ввода данных.
- **Устройства вывода**: Монитор, принтер, динамики и другие устройства для вывода информации.
- **Корпус**: Защищает внутренние компоненты и обеспечивает их охлаждение.

#### 2. Программное обеспечение (Software)
- **Операционная система (ОС)**: Управляет аппаратными ресурсами и предоставляет интерфейс для пользователя (например, Windows, macOS, Linux).
- **Прикладное программное обеспечение**: Программы, используемые для выполнения конкретных задач (например, текстовые редакторы, графические редакторы, браузеры).
- **Драйверы устройств**: Программное обеспечение, позволяющее ОС взаимодействовать с аппаратными компонентами (например, драйверы для принтеров, видеокарт).

#### 3. Сетевые компоненты
- **Сетевые адаптеры**: Устройства для подключения к локальным сетям и интернету (например, Wi-Fi адаптеры, Ethernet-карты).
- **Модем/роутер**: Устройства для подключения к интернету и распределения соединения между устройствами.

````python
import numpy as np

# Определение альтернатив
alternatives = {
    "A1": [11548, 57, 5, 1],
    "A2": [13385, 38, 3, 3],
    "A3": [18130, 59, 5, 5],
    "A4": [16921, 32, 5, 1],
    "A5": [12618, 41, 5, 2]
}

# Веса критериев
weights = np.array([0.4, 0.3, 0.2, 0.1])

# Преобразование состояния в числовые значения
# Плохое = 1, Среднее = 2, Хорошее = 3, Отличное = 5
# Это уже учтено в значениях альтернатив

# 1. Аддитивная свертка
def additive_aggregation(alternatives, weights):
    scores = {}
    for key, values in alternatives.items():
        score = sum(weights[i] * values[i] for i in range(len(weights)))
        scores[key] = score
    return scores

# 2. Мультипликативная свертка
def multiplicative_aggregation(alternatives, weights):
    scores = {}
    for key, values in alternatives.items():
        score = np.prod([values[i] ** weights[i] for i in range(len(weights))])
        scores[key] = score
    return scores

# 3. Метод идеальной точки
def ideal_point_method(alternatives, weights):
    ideal_point = [min([alt[i] for alt in alternatives.values()]) if i == 0 else max([alt[i] for alt in alternatives.values()]) for i in range(len(weights))]
    scores = {}
    for key, values in alternatives.items():
        score = sum(weights[i] * (ideal_point[i] - values[i]) if i == 0 else weights[i] * (values[i] - ideal_point[i]) for i in range(len(weights)))
        scores[key] = score
    return scores

# Выполнение методов
additive_scores = additive_aggregation(alternatives, weights)
multiplicative_scores = multiplicative_aggregation(alternatives, weights)
ideal_scores = ideal_point_method(alternatives, weights)

# Определение наиболее предпочтительных альтернатив
best_additive = max(additive_scores, key=additive_scores.get)
best_multiplicative = max(multiplicative_scores, key=multiplicative_scores.get)
best_ideal = min(ideal_scores, key=ideal_scores.get)

# Вывод результатов
print("Аддитивная свертка:", additive_scores)
print("Наиболее предпочтительная альтернатива (аддитивная):", best_additive)

print("\nМультипликативная свертка:", multiplicative_scores)
print("Наиболее предпочтительная альтернатива (мультипликативная):", best_multiplicative)

print("\nМетод идеальной точки:", ideal_scores)
print("Наиболее предпочтительная альтернатива (идеальная точка):", best_ideal)
````

**Ответ:**

Для оценки согласованности в методе анализа иерархий (AHP) необходимо рассчитать индекс согласованности (CI) и отношение согласованности (CR) для каждой матрицы парных сравнений. Вот шаги, которые нужно выполнить:

### Шаг 1: Расчет максимального собственного значения (λmax)

1. **Суммируем строки матрицы**: Для каждой строки матрицы парных сравнений суммируем значения.
2. **Умножаем на веса**: Умножаем каждую строку на соответствующий вес альтернативы.
3. **Суммируем результаты**: Суммируем полученные значения.
4. **Делим на сумму весов**: Делим результат на сумму весов, чтобы получить максимальное собственное значение (λmax) для матрицы.

### Шаг 2: Расчет индекса согласованности (CI)

Формула для расчета CI:
$$
CI = \frac{\lambda_{max} - n}{n - 1}
$$
где $ n $ — количество альтернатив.

### Шаг 3: Расчет отношения согласованности (CR)

Для расчета CR:
1. Определяем индекс случайной согласованности (RI) для $ n $ альтернатив. Для различных значений $ n $ RI составляет:
   - $ n = 2 $: RI = 0
   - $ n = 3 $: RI = 0.58
   - $ n = 4 $: RI = 0.90
   - $ n = 5 $: RI = 1.12
   - $ n = 6 $: RI = 1.24
   - $ n = 7 $: RI = 1.32

2. Используем формулу:
$$
CR = \frac{CI}{RI}
$$

## Запрос программного кода

Ниже представлен код на Python, который выполняет задачи, связанные с выбором парфюмерии с использованием метода анализа иерархий (AHP). Код включает в себя систематизацию данных, построение матриц парных сравнений, вычисление весов и оценку согласованности, а также формирование отчета.

````python

import numpy as np
import pandas as pd

# 1. Систематизация данных
criteria = [
    "Аромат",
    "Стоимость",
    "Стойкость",
    "Гипоаллергенность",
    "Известность бренда",
    "Дизайн флакона"
]

alternatives = {
    "A1": [5, 5, 2, 1, 1, 3],
    "A2": [4, 1, 4, 5, 3, 5],
    "A3": [5, 1, 5, 1, 4, 4],
    "A4": [1, 3, 3, 3, 2, 1],
    "A5": [1, 2, 2, 5, 3, 3],
    "A6": [4, 3, 3, 2, 3, 1]
}

# 2. Метод анализа иерархий (AHP)
def normalize_matrix(matrix):
    column_sums = matrix.sum(axis=0)
    normalized_matrix = matrix / column_sums
    return normalized_matrix

def calculate_weights(normalized_matrix):
    weights = normalized_matrix.mean(axis=1)
    return weights

def calculate_consistency_index(matrix, weights):
    n = matrix.shape[0]
    lambda_max = (matrix @ weights).sum() / weights.sum()
    ci = (lambda_max - n) / (n - 1)
    return ci, lambda_max

def calculate_cr(ci, n):
    # Random Index (RI) values for consistency check
    ri_values = {1: 0.00, 2: 0.00, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24}
    ri = ri_values.get(n, 1.24)  # Default to 1.24 for n > 5
    cr = ci / ri if ri != 0 else 0
    return cr

# Создание матриц парных сравнений для каждого критерия
def create_comparison_matrix(alternatives):
    n = len(alternatives)
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if i == j:
                matrix[i][j] = 1
            elif alternatives[i] > alternatives[j]:
                matrix[i][j] = 1 / (alternatives[i] / alternatives[j])
            else:
                matrix[i][j] = alternatives[i] / alternatives[j]
    return matrix

# Основной процесс
results = {}
for criterion in range(len(criteria)):
    comparison_matrix = create_comparison_matrix([alt[criterion] for alt in alternatives.values()])
    normalized_matrix = normalize_matrix(comparison_matrix)
    weights = calculate_weights(normalized_matrix)
    ci, lambda_max = calculate_consistency_index(comparison_matrix, weights)
    cr = calculate_cr(ci, len(alternatives))
    
    results[criteria[criterion]] = {
        "Weights": weights,
        "CI": ci,
        "CR": cr
    }

# Итоговые результаты
final_weights = np.zeros(len(alternatives))
for criterion in results:
    final_weights += results[criterion]["Weights"]

# Нормализация итоговых весов
final_weights /= np.sum(final_weights)

# Формирование отчета
report = {
    "Criteria": criteria,
    "Alternatives": alternatives,
    "Results": results,
    "Final Weights": final_weights
}

# Вывод отчета
print("Отчет по выбору парфюмерии")
print("\nКритерии:")
for criterion in criteria:
    print(f"- {criterion}")

print("\nАльтернативы:")
for alt, values in alternatives.items():
    print(f"{alt}: {values}")

print("\nПромежуточные результаты:")
for criterion, result in results.items():
    print(f"{criterion}:")
    print(f"  Веса: {result['Weights']}")
    print(f"  CI: {result['CI']:.4f}, CR: {result['CR']:.4f}")

print("\nИтоговые веса альтернатив:")
for i, alt in enumerate(alternatives.keys()):
    print(f"{alt}: {final_weights[i]:.4f}")

best_alternative = list(alternatives.keys())[np.argmax(final_weights)]
print(f"\nНаилучшая альтернатива: {best_alternative}")

````

2. **Метод анализа иерархий (AHP)**: 
   - Создаются матрицы парных сравнений для каждого критерия.
   - Нормализуются матрицы.
   - Вычисляются веса, индекс согласованности (CI) и отношение согласованности (CR).
3. **Формирование отчета**: 
   - Выводятся критерии, альтернативы и промежуточные результаты.
   - Рассчитываются итоговые веса и определяется наилучшая альтернатива.
